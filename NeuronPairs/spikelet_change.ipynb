# Linear function

```python
def spikelet_fun(strength):
    return strength* 50
```

```python
#loading up the parameters

params_HH = project.utils.load_HH()
params_LIF = project.utils.load_LIF()

#time parameters
t_final = 1500
dt = 0.01

#Other constant parameters
#Isyn = [[0, 0.5], [0.5, 0]]
Isyn_HH = [[0, 0.0], [0.0, 0]]
Isyn_IF = [[0, 0.0], [0.0, 0]]

Is_HH = np.array(Isyn_HH)
Is_IF = np.array(Isyn_IF)

k = 0.025

V0 = -65
tau = 0.65
y0 = [-65 , -65]
w0 = [0.0, 0.0]
u0 = [0,0]
n0 = 0
m0 = 0
h0 = 0
order = 2

#fixing random seed
np.random.seed(1234)

#generating the intensities
I_HH = np.zeros((int(t_final/dt),2))
I_IF = np.zeros((int(t_final/dt),2))

I_HH[:,0] = 2.5 
I_HH[:,1] = 0
I_IF[:,0] = 2.5 
I_IF[:,1] = 0
```

```python
dataHH_adjust, completeHH_adjust, matrixHH_adjust = project.models.HH_Neuron_Pairs(dt,t_final,order,[V0,V0],[n0,n0],[m0,m0],[h0,h0],*params_HH,I_HH,Is_HH,k,tau)
dataIF_adjust, completeIF_adjust, matrixIF_adjust = project.models.LIF_Neuron_Pairs(dt, t_final,order,y0,*params_LIF,I_IF,Is_IF,k,tau,spikelet_fun(k))
```

```python
time = np.linspace(0,t_final,int(t_final/dt),)
plt.plot(time,dataHH_adjust[:,0], color = 'red', label = 'HH')
plt.plot(time,dataIF_adjust[:,0], color = 'blue', label = 'IF')
plt.title(r'Spikelet Comparisson (I = 2$\mu$ A)')
plt.xlabel('Time (ms)')
plt.ylabel('Voltage (mV)')
plt.xlim(0,60)
plt.legend()
```

```python
time = np.linspace(0,t_final,int(t_final/dt),)
plt.plot(time,dataHH_adjust[:,1], color = 'red', label = 'HH')
plt.plot(time,dataIF_adjust[:,1], color = 'blue', label = 'IF')
plt.title(r'Post-Synaptic cells (I = 2$\mu$ A)')
plt.xlabel('Time (ms)')
plt.ylabel('Voltage (mV)')
plt.xlim(0,60)
plt.legend()
```

# Min and Maximum of the spikelet

```python
num_gap = 80
diff_HH = np.zeros(80)
diff_IF = np.zeros(80)

k = 0.005
k_plus = 0.0025
k = np.linspace(k,num_gap*k_plus + k,num_gap)

for i in range(0,num_gap):
    dataHH_adjust, completeHH_adjust, matrixHH_adjust = project.models.HH_Neuron_Pairs(dt,t_final,order,[V0,V0],[n0,n0],[m0,m0],[h0,h0],*params_HH,I_HH,Is_HH,k[i],tau)
    dataIF_adjust, completeIF_adjust, matrixIF_adjust = project.models.LIF_Neuron_Pairs(dt, t_final,order,y0,*params_LIF,I_IF,Is_IF,k[i],tau,spikelet_fun(k[i]))

    diff_HH[i]= np.max(dataHH_adjust[10000:,1]) - np.min(dataHH_adjust[10000:,1])
    diff_IF[i] = np.max(dataIF_adjust[10000:,1]) - np.min(dataIF_adjust[10000:,1])
```

```python
plt.plot(k,diff_HH, label = 'HH')
plt.plot(k,diff_IF,label = 'IF')

plt.legend()
```

```python
import scipy
def fit_fun (x,a,b):
    return x*a + b
popt, pcov = scipy.optimize.curve_fit(fit_fun,k,diff_HH)
```

```python
def fit_fun_2 (x,a):
    return np.exp(x) * a *x
popt_2, pcov_2 = scipy.optimize.curve_fit(fit_fun_2,k,diff_HH)
```
